Code for Feature Extraction from VGG16 Network

```python
import numpy as np
import skimage.io
import skimage
from skimage.color import rgb2hsv
from os.path import dirname
from os.path import basename
from tqdm import tqdm
import glob
import pickle
import os
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import matplotlib.ticker as ticker
from matplotlib.cm import ScalarMappable
sns.set_style("white")

sns.set_context("talk", font_scale = 1.3)
matplotlib.rcParams['mathtext.fontset'] = 'stix'
matplotlib.rcParams['font.family'] = 'STIXGeneral'
matplotlib.rcParams['font.size'] = 20
matplotlib.rcParams['axes.axisbelow'] = True
matplotlib.rcParams.update({'figure.autolayout': True})
matplotlib.rcParams['figure.figsize'] = 9,7
```

Identify path to RGB images that are to be analyzed.

```python
path_list = glob.glob('LC_Response_Data/*/10mMAl*.mpg00*Row00*Col00*.jpg')
```

hue histogram from each image in our dataset

```python
labels = []
Hue_Descriptors_tot = []

for y in tqdm(path_list):    
    
    #Load and preprocess images.
    img = skimage.io.imread(y)
    
    Hue = rgb2hsv(img)[:,:,0]
    
    bins = np.arange(0,1,.01)
    
    Hue_Hist = np.histogram(Hue.flatten(),bins = bins)[0]
    
    #Store label of each image.
    label = basename(dirname(y))
    labels = np.append(labels,label)
    
    #Create large dictionary containing all descriptors.        
    Hue_Descriptors_tot.append(Hue_Hist[:])     
    
```

Plot Example Hue Histogram

```python

bins = np.arange(0,1,.01)

Hue_Hist = np.histogram(Hue.flatten(),bins = bins)[0]

fig, ax = plt.subplots()


plt.bar(bins[2::],Hue_Hist[1::]/(np.sum(Hue_Hist[1::])),width = .01)

my_cmap = plt.cm.get_cmap('hsv')
sm = ScalarMappable(cmap=my_cmap, norm=plt.Normalize(0,1))
sm.set_array([])

cbar = plt.colorbar(sm,orientation = "horizontal",pad = 0.05)
cbar.set_label('Hue', rotation=0,labelpad=25)

plt.xticks([])
plt.yticks([0,.1,.2,.3,.4])
plt.ylabel("Probability")


plt.savefig("Hue_Hist_Water.eps")
```

```python

img = skimage.io.imread(path_list[0])

Hue = rgb2hsv(img)[:,:,0]

bins = np.arange(0,1,.01)

Hue_Hist = np.histogram(Hue.flatten(),bins = bins)[0]

fig, ax = plt.subplots()

plt.bar(bins[2::],Hue_Hist[1::]/(np.sum(Hue_Hist[1::])),width = .01)

my_cmap = plt.cm.get_cmap('hsv')
sm = ScalarMappable(cmap=my_cmap, norm=plt.Normalize(0,1))
sm.set_array([])

cbar = plt.colorbar(sm,orientation = "horizontal",pad = 0.05)
cbar.set_label('Hue', rotation=0,labelpad=25)

plt.xticks([])
plt.yticks([0,.1,.2,.3,.4])
plt.ylabel("Probability")


plt.savefig("Hue_Hist_DMMP.eps")
```

Plot Cumulative Distribution Plots for Hue Histograms

```python


img = skimage.io.imread(path_list[0])
Hue = rgb2hsv(img)[:,:,0]
bins = np.arange(0,1,.01)
Hue_Hist = np.histogram(Hue.flatten(),bins = bins)[0]
huesum = np.cumsum(Hue_Hist[1::]/np.sum(Hue_Hist[1::]))


sns.lineplot(data = huesum,label = "DMMP",size = 5)


img = skimage.io.imread(path_list[-1])
Hue = rgb2hsv(img)[:,:,0]
bins = np.arange(0,1,.01)
Hue_Hist = np.histogram(Hue.flatten(),bins = bins)[0]
huesum = np.cumsum(Hue_Hist[1::]/np.sum(Hue_Hist[1::]))

sns.lineplot(data = huesum,label = "Water",size = 10)

my_cmap = plt.cm.get_cmap('hsv')
sm = ScalarMappable(cmap=my_cmap, norm=plt.Normalize(0,1))
sm.set_array([])

cbar = plt.colorbar(sm,orientation = "horizontal",pad = 0.05)
cbar.set_label('Hue', rotation=0,labelpad=25)

plt.xticks([])
plt.ylabel("Cumulative Distribution")

plt.savefig("Cum_Dist_Hue.eps")
```

Store information in Pickle File

```python
with open('Pickle_Files/Hue_Descriptors','wb') as fp:
    pickle.dump(Hue_Descriptors_tot,fp)
    
with open('Pickle_Files/Hue_Labels','wb') as fp:
    pickle.dump(labels,fp)   
```
