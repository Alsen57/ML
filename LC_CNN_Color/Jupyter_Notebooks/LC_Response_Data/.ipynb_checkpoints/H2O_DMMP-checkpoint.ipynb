```python
def Stack_images(path_loc,time_min,time_max,mode):

    flats = []
    
    path = glob.glob(path_loc)
    
    Times = []
    
    data = []
    
    dims = []
    #compile Times
    
    for i in path:
    
        Time = np.int(os.path.basename(i)[-9:-4])
        
        Times.append(Time)
      
    Times = np.hstack(Times)
       
    #Time Frame    
    
    a = np.where((Times > time_min) & (Times < time_max))[0].astype(int)
    
    Pathtime = [path[i] for i in a]

    if mode == 0:
    
        for j in Pathtime:
            
            image = skimage.io.imread(j)
            grayimage = skimage.color.rgb2gray(image)
#             flatimage = np.ndarray.flatten(grayimage)
#             low_values_flags = flatimage < .1  # Where values are low
#             flatimage[low_values_flags] = 0
            
            grayimage = (grayimage - np.min(grayimage))/(np.max(grayimage) - np.min(grayimage))
    
            flats.append(grayimage)
            
            dims.append([grayimage.shape[0],grayimage.shape[1]])
            
            data = flats
        
    else:
        
        for j in Pathtime:
            
            image = skimage.io.imread(j)
            hueimage = skimage.color.rgb2hsv(image)[:,:,0]
#             flatimage = np.ndarray.flatten(hueimage)
            flats.append(hueimage)
            
            dims.append([hueimage.shape[0],hueimage.shape[1]])
             
            data = flats
    
    return data,dims
```

```python
data_30,dims_30 = Stack_images(r"C:\Users\adsmith23\ML\LC_CNN_Color\Jupyter_Notebooks\LC_Response_Data\DMMP\*.jpg",2000,2020,0)

data_50,dims_50 = Stack_images(r"C:\Users\adsmith23\ML\LC_CNN_Color\Jupyter_Notebooks\LC_Response_Data\DMMP\*.jpg",2000,2020,0)
```

```python
os.makedirs("pd", exist_ok=True)  # 保存先のディレクトリを作成
for i in range(96):
    hc.PDList.from_bitmap_levelset(data_30[i], save_to="pd{}.idiagram".format(i))
```

```python
pds = [hc.PDList("pd{}.idiagram".format(i)).dth_diagram(1) for i in range(96)]
```

```python
mesh = hc.PIVectorizerMesh((0,1), 128, sigma=0.02, weight=("linear",1))
```

```python
pdvects30 = np.vstack([mesh.vectorize(pd) for pd in pds])
```

```python
os.makedirs("pd1", exist_ok=True)  
for i in range(96):
    hc.PDList.from_bitmap_levelset(data_50[i],save_to="pd1{}.idiagram".format(i))
```

```python
pd1s = [hc.PDList("pd1{}.idiagram".format(i)).dth_diagram(1) for i in range(96)]
```

```python
pdvects50 = np.vstack([mesh.vectorize(pd) for pd in pd1s])
```

```python
pdtot = np.vstack([pdvects30,pdvects50])
pdtot = pdtot / pdtot.max()
```

```python
label = np.zeros(192)
label[0:96] = 1
```

```python
pca = PCA()
X = pca.fit_transform(pdtot)
```

```python
plt.scatter(X[0:96,0],X[0:96,2],label = "Low Conc.")
plt.scatter(X[96::,0],X[96::,2],label = "High Conc.")
plt.legend(fontsize = 20)
plt.xlabel('Principal Component 1')

plt.ylabel('Principal Component 2')
plt.savefig("PCA_lcs.eps")
plt.savefig("PCA_lcs.svg")
```

```python

mesh.histogram_from_vector(pca.components_[3, :]).plot(colorbar={"type":"linear-midpoint", "midpoint": 0})
plt.savefig("pdweights.svg",transparent = True)
```

```python
coef = pca.components_[1,:]
```

```python
pdvects_train, pdvects_test, labels_train, labels_test = train_test_split(pdtot, label, test_size=0.3)
```

```python
model = lm.LogisticRegression(C=1)
```

```python
model.fit(pdvects_train, labels_train)
```

```python
model.score(pdvects_test, labels_test)
```

```python

import matplotlib
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import numpy as np
class MidpointNormalize(colors.Normalize):
	"""
	Normalise the colorbar so that diverging bars work there way either side from a prescribed midpoint value)

	e.g. im=ax1.imshow(array, norm=MidpointNormalize(midpoint=0.,vmin=-100, vmax=100))
	"""
	def __init__(self, vmin=None, vmax=None, midpoint=None, clip=False):
		self.midpoint = midpoint
		colors.Normalize.__init__(self, vmin, vmax, clip)

	def __call__(self, value, clip=None):
		# I'm ignoring masked values and all kinds of edge cases to make a
		# simple example...
		x, y = [self.vmin, self.midpoint, self.vmax], [0, 0.5, 1]
		return np.ma.masked_array(np.interp(value, x, y), np.isnan(value))
```

```python

red_area = mesh.mask_from_vector(coef > 0.09)
blue_area = mesh.mask_from_vector(coef < -0.09)

fig = plt.figure()
a = mesh.histogram_from_vector(coef + .000001).values*100

a[a==0] = np.nan

elev_min = .1*100
elev_max = -.1*100

# a[(a>=-1e-06) & (a<=1e-2)] = np.nan


mask = np.zeros_like(a)

mask[np.triu_indices_from(mask)] = True

sns.axes_style("darkgrid")

plt.plot([0,400],[0,400])

sns.heatmap(a,cmap = "coolwarm",mask=mask,clim=(elev_min, elev_max), norm=MidpointNormalize(midpoint=0,vmin=elev_min, vmax=elev_max),cbar_kws={'label': 'Classifier Weights'},square = True)
plt.grid(True)
plt.xticks(np.linspace(0,120,num=5))
plt.yticks(np.linspace(0,120,num=5))
plt.gca().yaxis.set_major_formatter(StrMethodFormatter('{x:,.1f}')) # 2 decimal places
plt.gca().xaxis.set_major_formatter(StrMethodFormatter('{x:,.1f}')) # 2 decimal places

plt.gca().invert_yaxis()
plt.gca().axes.get_xaxis().set_ticklabels([])
plt.gca().axes.get_yaxis().set_ticklabels([])
plt.ylabel("Death")
plt.xlabel("Birth")

plt.savefig("so2weights.pdf")
```
